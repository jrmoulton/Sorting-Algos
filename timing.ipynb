```python
import pandas as pd
import numpy as np
import enum
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
class Algo(enum.Enum):
   Bubble = 0
   Selection = 1
   Insertion = 2
   Quick = 3
```

```python
data = pd.read_csv("time3.csv")
data.columns =["algo", "size", "time"]
```

```python
data
```

```python
fig, ax = plt.subplots()
font = {'family' : 'Helvetica',
        'weight' : 'bold',
        'size'   : 17}

matplotlib.rc('font', **font)
data.groupby("algo").plot(x="size", y="time", ax=ax, figsize=(16,8))
ax.set_ylabel("Seconds")
ax.set_xlabel("Elements in array")
ax.legend(Algo);
```

```python
data.groupby("algo")["time"].mean().rename(index=Algo)
```
